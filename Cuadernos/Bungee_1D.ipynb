```python
from Solver_EDOS import RK38,np # Solver de EDOS
import matplotlib.pyplot as plt # graficar

###
g=9.8
c_d=0.25
m=68.1

## Modelo

F= lambda t,X:np.array([-X[1],
                        g-c_d/m*X[1]**2])

F1=lambda t,X:np.array([-X[1],
                        g-c_d/m*X[1]*abs(X[1])])

### Con in
X0=[0,-40]

#tiempo
T=[0,12]
#par num
n=1000

t,X=RK38(F,X0,T,n)
t,X1=RK38(F1,X0,T,n)
```

```python
plt.figure(figsize=(18,18))

plt.suptitle("Modelo Bungee",fontsize=15)
plt.subplot(221)
plt.plot(t,X[:,0],'r--',label="Posicin, Incorrect Drag")
plt.plot(t,X1[:,0],'g',label="Posicin, Correct Drag")
plt.title("Posicin vs Tiempo",fontsize=15)
plt.xlabel("$t$",fontsize=15)
plt.ylabel("$x$",fontsize=15)
plt.legend()
plt.grid()

plt.subplot(222)
plt.plot(t,X[:,1],'r--',label="Posicin, Incorrect Drag")
plt.plot(t,X1[:,1],'g',label="Posicin, Correct Drag")
plt.title("Velocidad vs Tiempo",fontsize=15)
plt.xlabel("$t$",fontsize=15)
plt.ylabel("$v$",fontsize=15)
plt.grid()
plt.legend()

plt.subplot(2,2,(3,4))
plt.plot(X[:,0],X[:,1],'r--',label="Incorrect Drag")
plt.plot(X1[:,0],X1[:,1],'g',label="Correct Drag")
plt.xlabel("$x$",fontsize=15)
plt.ylabel("$v$",fontsize=15)
plt.title("Espacio Fase",fontsize=15)
plt.legend()
plt.grid()

plt.show()
```
